```python
# imports the libaries
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

# load the iris dataset
iris = load_iris()

# create a Gaussian Naive Bayes model
gnb = GaussianNB()

# perform a 5-fold cross validation
scores = cross_val_score(gnb, iris.data, iris.target, cv=5, scoring='accuracy')

# print the scores
print("scores of each fold", scores)
print("average score", scores.mean())
print("standard deviation of scores", scores.std())

# output
# scores of each fold [0.93333333 0.96666667 0.93333333 0.96666667 1.        ]
```

```python
# plot k-fold CV
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
# load the dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# define the decision tree classifier
clf = DecisionTreeClassifier()

# preform k-fold CV with k=5
scores = cross_val_score(clf, X,y, cv=8)

# plot the results
plt.figure(figsize=(6,4))
plt.plot(range(1,9), scores, color = 'blue', linestyle='dashed',marker='o', mfc='red',ms=10)
plt.title("k-fold cv results")
plt.xlabel("fold number")
plt.ylabel("Accuracy")
plt.ylim([0.5,1.0])
plt.show()
```
