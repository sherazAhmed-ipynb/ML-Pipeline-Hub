```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder,OrdinalEncoder
```

```python
# laoding the dataset
df = sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

```python
df['day'].value_counts()
```

```python
# encode the time columns with sk_learn
le = LabelEncoder()

df['encodedd_time'] = le.fit_transform(df['time'])
df.head()
```

```python
df['encodedd_time'].value_counts()
```

```python
df['day'].value_counts()
```

# ordinal encoding 

```python
oe = OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
```

```python
df['encoded_days'] = oe.fit_transform(df[['day']])
df.head(10)
```

```python
df['day'].value_counts()
```

```python
df['encoded_days'].value_counts()
```

# one hot encoding

```python
ohe = OneHotEncoder()
ohe.fit_transform(df[['day']]).toarray()
```
